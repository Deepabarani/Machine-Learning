```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
#Order data - size (Ordinal)
data = {
    "Item": ['shirt','t-shirt', 'skirt', 'jeans'],
  "Color": ['blue', 'green', 'pink', 'yellow'],
    'Size': ['xl', 'xxl', 's', 'm']
}
```

```python
df = pd.DataFrame(data)
df
```

```python
encoder = LabelEncoder()
```

```python
df['Encoded_Size'] = encoder.fit_transform(df['Size'])
```

```python
df
```

```python
#unordered data (OneHotEncoder)
dataFruits = {
    "fruits": ['apple', 'banana', 'grapes'],
    "color": ["red", "yellow", "purple"]
}
```

```python
df = pd.DataFrame(dataFruits)
```

```python
encoder= OneHotEncoder()
```

```python
enc_colors = encoder.fit_transform(df[['color']]).toarray()
```

```python
enc_df = pd.DataFrame(enc_colors, columns=encoder.get_feature_names_out(['color']))
```

```python
final_encoded = pd.concat([df, enc_df], axis=1)
final_encoded
```
